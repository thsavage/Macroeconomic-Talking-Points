# Three Empirical Regularities Relevant for CRE

1. Long-duration rates, such as the 10-year, are falling.
2. Pre-pandemic, cap rates were compressing.
     * Due to current monetary policy, they are rising.
3. Correlation between short- and long-duration rates has returned 
    * Refuting Bernanke's conjecture of a savings glut.

## June 2020
* Q2 2020 GDP contraction much larger than global financial crisis.
* 10-year U.S. Treasury had fallen to ~50 bps.
* NBER formally declares the U.S. to be in a recession. 

## The Current Economic Cycle (2021 - ?)

* **The U.S. is not in a recession based on labor market performance and strong real growth**.
    * The first wave of inflation was driven by distortions in consumption and supply-chain issues.  
    * The second wave of inflation was driven by Russia's invasion of Ukraine.  
    * The thrid wave of inflation results from firms taking potential advantage of some price-setting power.
